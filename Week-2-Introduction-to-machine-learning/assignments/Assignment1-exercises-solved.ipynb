```python
# 2.1. Write a for loop that prints out all integers in the range 0 to 1000 inclusive that are divisible by both 3 and 5
for i in range(1001):
    if i % 3 == 0 and i % 5 == 0:
        print(i, end=' ')

# 2.2. Write a function that takes as input a list of numbers and returns the product of all the numbers in the list
#    You can assume that every element of the list is an int or a float
def product(number_list):
    result = 1
    for x in number_list:
        result = result * x
    return result

# Test the product function
print(product([2, 4, 6, 8, 10]))

# 3.3. Write a function that takes a positive integer >= 2 as input and returns `True` if the number is prime and `False` otherwise.
# A number p is prime if it is divisible only by itself and 1, i.e. `p % n != 0` for all n >= 2 and < p.
# You can assume that the input value is a positive integer >= 2.
def is_prime(n):
    for i in range(2, n): 
        if n % i == 0:
            return False
    return True

# Test the is_prime function
print(is_prime(71))
print(is_prime(48))
```
